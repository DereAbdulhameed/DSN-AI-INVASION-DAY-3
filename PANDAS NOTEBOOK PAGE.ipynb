# INTRODUCTION TO PANDAS

Pandas is basically used to read all kinds of including CSV files, excel files, JASON files. This is because it has more flexibilitu and allows you to work with larger datasets.

```python
import pandas as pd
```

```python
# Ways to manually create a dataframe

# Initialize list of lists
data = [['Ayo', 10],['Imran', 15],['Chucks', 14]]

# create the pandas Dataframe from the list and adding the column headers
df = pd.DataFrame(data, columns = ['Name','Age'])

# print the dataframe
df
```

# DESCRIBING / SORTING DATA

#### Describe function is used to carry out some statistical functions on the dataframe. Describe functions works only on integer and float values BUT does not work on the object values i.e string values

```python
df.describe()
```

```python
# To include summary of descriptive statistics of non numeric columns of the dataframe, we use the notation (include = 'all')
df.describe(include = 'all')
```

Other descriptive statistics functions are:
1. count()Number of non null observations

2. sum()Sum of values

3. mean()Mean of Values

4. median()Median of Values

5. mode()Mode of values

6. std() Standard Deviation of the Values

7. min() Minimum Value

8. max()

9. Maximum Value

10. abs()Absolute Value

11. prod() Product of Values

12. cumsum()Cumulative Sum

13. cumprod()Cumulative Product

```python
df['Population'].mean()
```

#### Sort function is used to arrange the datas in the dataframe ina chronological order, either in an ascending format or a descending format.

```python
df.sort_values('STATES', ascending=False)
```

# PANDAS INDEXING

There are several ways of indexing a Pandas DataFrame. it include

a) Square bracket notation


b) loc and iloc: loc is label-based, which means that you have to specify rows and columns based on their row and column labels WHILE iloc is integer index based, so you have to specify rows and columns by their integer index dot(.) notation

```python
# TO READ A SINGLE ROW I.E AN INDEX
df.iloc[2]
```

```python
df.iloc[3:5]
```

```python
# To read a specified location. The format is [R,C]
df.iloc[2,2]
```

```python
# Changing the dtype of features for series object
df['Population'] = df['Population'].astype('float')

df
```
